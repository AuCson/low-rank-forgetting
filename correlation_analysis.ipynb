## Sentence Embedding similarity

```python
cos_mat_sentencebert = np.nan_to_num(get_cos_dist_mat_direct('sentencebert'))
cos_mat_openai = np.nan_to_num(get_cos_dist_mat_direct('openai'))
```

```python
print(cos_mat_openai.shape, cos_mat_sentencebert.shape)
```

```python
# evaluate on some random subset for faster computation
rand_idxs = np.sort(deterministic_random_indices(141871, 10000))
```

```python
fgt_arr_1b, meta = load_concat_arr('stats/olmo-1b/ftd-ppl-inc-full.npy')
# 0.04% of data can have nan fgt values
fgt_arr_1b = np.nan_to_num(fgt_arr_1b)
fgt_arr_1b_subset = fgt_arr_1b[:,rand_idxs]
```

```python
fgt_arr_7b, meta = load_concat_arr('stats/olmo-7b/ftd-ppl-inc-full.npy')
fgt_arr_7b = np.nan_to_num(fgt_arr_7b)
fgt_arr_7b_subset = fgt_arr_7b[:,rand_idxs]
```

```python
def p_s_corr(x, y):
    print('Pearsonr:', pearsonr(x, y)[0], 'Spearmanr:', spearmanr(x, y)[0])
```

```python
# OLMo-7B + OpenAI
p_s_corr(fgt_arr_7b_subset.flatten(), cos_mat_openai[:,rand_idxs].flatten())
```

```python
# OLMo-7B + OpenAI
p_s_corr(fgt_arr_7b_subset.flatten(), cos_mat_openai[:,rand_idxs].flatten())
```

```python
# OLMo-7B + SentenceBERT
p_s_corr(fgt_arr_7b_subset.flatten(), cos_mat_sentencebert[:,rand_idxs].flatten())
```

```python
# OLMo-1B + OpenAI
p_s_corr(fgt_arr_1b_subset.flatten(), cos_mat_openai[:,rand_idxs].flatten())
```
